Test for creating graphs using networkx, and matplotlib for generating graphs. This program follows a cop game of Cops and Robbers.
TO DO: The robber in this program is immobile and stay on current node, this should be changed. 

```python
import networkx as nx
import matplotlib.pyplot as plt
import random
```

```python
# Create a cycle based on the number of nodes
num_nodes = 8
G = nx.cycle_graph(num_nodes)
```

```python
# Robbers and Cops are placed onto different nodes
cops = random.sample(range(num_nodes), 2)  # Places 2 cops
robber = random.choice([node for node in G.nodes() if node not in cops]) #robber is distinctly placed on a vertex not cop
```

```python
# Moves cops to an adjacent vertex
def move_cops():
    for i, cop in enumerate(cops):
        neighbors = list(G.neighbors(cop))
        if robber in neighbors:
            # If the robber is a neighbor, move towards it
            next_pos = robber
        else:
            # Move randomly to one of the neighbors
            next_pos = random.choice(neighbors)
        cops[i] = next_pos

```

```python
# When cops catches the robber
def is_caught():
    return robber in cops
```

```python
# Plot initial positions of cops and robbers
pos = nx.circular_layout(G)
nx.draw_networkx_nodes(G, pos, nodelist=cops, node_color='blue', label='Cops')
nx.draw_networkx_nodes(G, pos, nodelist=[robber], node_color='red', label='Robber')
nx.draw_networkx_edges(G, pos)
plt.legend()
plt.title("Initial Positions")
plt.show()
```

```python
# Move the cops until the robber is caught
while not is_caught():
    move_cops()
```

```python

# Plot final positions
nx.draw_networkx_nodes(G, pos, nodelist=cops, node_color='blue', label='Cops')
nx.draw_networkx_nodes(G, pos, nodelist=[robber], node_color='red', label='Robber (Caught)')
nx.draw_networkx_edges(G, pos)
plt.legend()
plt.title("Final Positions (Robber Caught)")
plt.show()
```
