```python
def get_positive_int(question):
    '''
    Repeatedly prompts the user for a positive integer
    till one is given, then returns this value.
    Uses the given "question" as the prompt to user.
    '''
    while True:
        try:
            val = int(input(question + ":"))
            if val <= 0:
                print("Must be positive!")
            else:
                break
        except:
            print("Give me a valid integer!")
            
    return val
```
