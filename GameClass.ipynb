# The Game Class

Has the Game class.  And a create_game function below.

```python
class Game:

    def __init__(self, graph, numCops): 
        '''
        Input:graph which is a Graph object, C which is used to place cop, R which is used to place robber
        Result: initializes the game, create inner variables board, turn, numbherOfTurns, copPosition, and robberPositions
        NOTE: THE USER INTERFACE SHOULD LET THE COPS PLACE FIRST, THEN THE ROBBER,
        '''
        self.board = graph
        self.whoTurn = "cop" # can be value "cop" or "rob"
        self.numTurns = 10   # defaults to 10 but can be changed
        self.numCops  = 0   # Zero unless reset
        
        self.numCops = numCops
    

    def get_rob_loc(self):
        return self.rob
    
    def get_cop_loc(self):
        return self.cops    

    def get_nodes(self):
        return self.board.nodes
    
    def get_num_cops(self):
        if type(self.numCops) == 'list': 
            return len(self.numCops)
        
        return self.numCops


    def place_cop(self, coplist):
        '''
        Input: A list of cop positions representing new coordinates for each cop, (This assumes that all the moves are legal)
        Output: True if legal placement; and False if not
        ONLY USE THIS WHEN CREATING GAME CLASS
        '''
        for i in coplist:
            if not i in self.board.nodes:
                raise "bad"

        self.cops = coplist
        self.numCops = len(self.cops)
        return True
        

    def place_rob(self, robber):
        '''
        input: An integer representing a vertice(assumes it is a legal move)
        output:
        does not return anything
        ONLY USE THIS WHEN CREATING GAME
        '''
        
        if robber not in self.board.nodes:
            raise "bad" 
            
        self.rob = robber
        return True

    
    def get_moves(self, v):
        if v not in self.board.nodes():
            return None
        
        moves = list(self.board.neighbors(v))
        moves.append(v)
        return moves

    
    def get_which_player_turn(self):
        '''
        Input:none
        Result: returns turn which will be either "cop" or "rob"
        '''
        return self.turn


    def get_turn_count(self):
        '''
        Input:none
        Result:returns NumberOfTurns
        '''
        return self.numTurns
    
   
    def make_rob_move(self, newRobberPosition):
        '''
        Input: takes n as an input, n should be a node adjacent to robber position
        Result: robPosition is changed to newRobberposition2 and turn is change to "cop"
                if it is a legal move.  And in that case returns True.
                if illegal move, turn stays robbers, and no position
                changed and return False
        '''
        if newRobberPosition in self.get_moves(self.rob):
            self.rob = newRobberPosition
            self.turn = "cop" #changes to cop turn
            self.numTurns -= 1
            return True
        else:
            return False
        
   
    def make_cop_move(self, listOfCops):
        '''
        Input: takes list of cops as input
        Result: If all moves are legal then is changed and turn is 
                change to "rob".  And in that case returns True.
                if illegal move, turn stays cop turn, and no position
                changed and return False
        '''
        
        newPositions = []
        for copNum in range(self.numCops):
            print(copNum, listOfCops)##############################
            current = listOfCops[copNum]
            if current in self.get_moves(self.cops[copNum]):
                newPositions.append(current)
            else:
                return False
        
        # so all the moves were good
        self.cops = newPositions
        self.turn = "rob" # changes to rob turn
        return True

    
    def determine_state(self):
        '''
        Returns as a string, possible states of game:
        - 'CopWin': the cops have won
        - 'TurnMax': the robber has won because turns run out
        - 'Neutral': neither has won
        '''
        if self.rob in self.cops:
            return('CopWin')
        if self.numTurns <= 0:
            return('TurnMax')
        else: 
            return('Neutral')
        
        
    def display_game_state(self):
        'Returns a string of nicely displayed game state'
        state = self.determine_state()
        if state == 'CopWin':
            return "The cops have won!"
        elif state == 'TurnMax':
            return "The robber won! (maximum turns reached)"
        else:
            return "No winner yet ..."

    
    def change_num_turns(self, n):
        '''
        Input: Single positive integer (no error checking)
        Result: No output, but makes the number of turns change to input n
        '''
        self.numTurns = n 

        
    def display_graph(self):
        '''
        Input: none 
        Result: Should print out a list of nodes and edges
        '''
        
        #Prints text output of the graph
        print( 'Vertices are:', self.board.nodes)
        print('Edges are:', self.board.edges)
    
    
    def display_game(self):
        '''
        Input: none 
        Result: Should print out state of the game including the graph
        '''
        
        # Prints text output of the graph
        self.display_graph()
        print("Number of Turns Left:", self.numTurns)
        print(self.display_game_state())
        
        c = 1
        for i in self.cops:
            print('cop '+ str(c) +' is at: ' + str(i))
            c += 1
            
        print('The robber is at: '+ str(self.rob))
        
        
    def play_with_printing(self, cop_strat, rob_strat):
        # COMMENT: It is verbose.
        # After play returns some stats:
        # - Who won
        # - Number of turns used.
        # - The history of play (if verbose = True)
        
        cop_start = cop_strat.placePieces(self)
        self.place_cop(cop_start)
        
        rob_start = rob_strat.placePieces(self)
        self.place_rob(rob_start)
        
        
        try:###################################################
            print(cop_strat.flowdict)
        except:
            pass
        print("*** GAME START ***")
        self.display_game()
        print("*******")
        print()
        
        while True:
            try:#######################################
                print(cop_strat.flowdict)
                print(cop_strat.movelist)
            except:
                pass
            if self.determine_state() != 'Neutral': #Fix to the game not ending right away
                break
            cop_list = cop_strat.makeMove(self)
            self.make_cop_move(cop_list)
            print("* Cops Just Moved *")
            self.display_game()
            print()
            if self.determine_state() != 'Neutral':
                break
            
            rob_pos = rob_strat.makeMove(self)
            self.make_rob_move(rob_pos)
            print("* Robber Just Moved *")
            self.display_game()
            print()
            if self.determine_state() != 'Neutral':
                break
            
        print()
        print()
        print("*** GAME ENDED IN THE FOLLOWING STATE ***")
        self.display_game()
        print("************")

```

# Comments on create_game function

create_game() function, its parameters are the same as creating a 
Game object. The main difference is that if a game is not created with the right parameters, creation will fail

instead of       game1=Game(g,Cops,rob)
use the format   game1=create_game(g,Cops,rob)

if creation happens then game1 will be a regular game object
if it fails, then it will NOT be a game object

```python
def create_game(graph, coplist, robber):
    try: 
        game = Game(graph, coplist, robber)
        
        return game
    except:
        print("failure to create game with," , "graph:", graph, "cop positions:", coplist, "robber:", robber)
        return 
```
